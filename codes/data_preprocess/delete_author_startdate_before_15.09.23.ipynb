```python
import pandas as pd
first = pd.read_csv('author_firstdate_depression.csv')

#첫 글이 해당 시간 이전인 사람의 리스트인 delete_f 만들기
f= first[first['first_date'] <'2015-09-23 00:00:00'] 
f = f.drop(f.columns[0],axis='columns')
delete_f = f['author'].values
list(delete_f)
```

```python
before = pd.read_csv('depression_author_before.csv')
after = pd.read_csv('depression_author_after.csv')

#delete_f에 있는 사람 빼고 나머지를 다시 저장하는 과정
mask = before['author'].isin(list(delete_f))
before[~mask].drop(before[~mask].columns[[0]],axis = 'columns').to_csv('depression_author_before_del.csv',index=True)

mask = after['author'].isin(list(delete_f))
after[~mask].drop(after[~mask].columns[[0]],axis='columns').to_csv('depression_author_after_del.csv',index=True)
```
