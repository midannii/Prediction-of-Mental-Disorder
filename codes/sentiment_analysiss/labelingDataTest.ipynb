```python
def cleansing(content):
    
    cleasing_content = []
    url_pattern ='https?://\S+|#([0-9a-zA-Z]*)'

    for text in content:
        #대문자 소문자 변환
        lower_content = (text.lower())

        #불용어 제거
        shortword = re.compile(r'\W*\b\w{1,2}\b')
        shortword_content = shortword.sub('', lower_content)
        text = re.sub(pattern=url_pattern, repl=' ', string = shortword_content)
        
        text = re.sub('[\"@,\[\]\(\)&%*-=#/:$\}\{”]', '', text)
        text = text.replace("\n", "")
        text = text.replace("\r", "")
        cleasing_content.append(text)
    return cleasing_content
```

```python
import sys, os
from labMTsimple.storyLab import *
import codecs ## handle utf8
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1,1))

emotionFileReader(stopval=0.0,lang='english',returnVector=True)
labMT,labMTvector,labMTwordList = emotionFileReader(stopval=0.0,lang='english',returnVector=True)
def my_labMT(text):
    labMT,labMTvector,labMTwordList = emotionFileReader(stopval=0.0,lang='english',returnVector=True)
    ## compute valence score and return frequency vector for generating wordshift
    Valence,Fvec = emotion(text,labMT,shift=True,happsList=labMTvector)
    ## but we didn't apply a lens yet, so stop the vectors first
    StoppedVec = stopper(Fvec,labMTvector,labMTwordList,stopVal=1.0)
    Valence = emotionV(StoppedVec,labMTvector)
    #print('the valence of {0} is {1:.5}'.format('input data',Valence))
    return Valence
```

```python
def s_cleansing(List):
    
    corpus = []
    for i in List:
        url_pattern ='https?://\S+|#([0-9a-zA-Z]*)'
        i = re.sub(pattern=url_pattern, repl=' ', string = str(i))
        i = i.replace('\r','').replace('\n','')
        text = re.sub('[-=+,#/\?:^$”@*\"“※~&%ㆍ!』\\‘|\(\)\[\]\<\>`…》;’]', '', i.lower())
        corpus.append(text)
    return corpus
```

# 글 전체

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
from afinn import Afinn
import time
start = time.time()
afn = Afinn(emoticons=True)
v_label = []
a_label = []
t_label = []

num = 1
sid = SentimentIntensityAnalyzer()
for i in sample_cleansing:
    if sid.polarity_scores(i)['compound'] > 0:
        v_label.append(1)
    else:
        v_label.append(0)
    if TextBlob(i).sentiment.polarity > 0:
        t_label.append(1)
    else:
        t_label.append(0)
    if afn.score(i) > 0:
        a_label.append(1)
    else:
        a_label.append(0)
print(time.time()-start)
```

```python
l_label = []
for i in sample_cleansing:
    if my_labMT(i) > 5.5:
        l_label.append(1)
    else:
        l_label.append(0)
```

```python
num_v_all = 0
num_t_all = 0
num_a_all = 0
num_l_all = 0
for i in range(len(sample)):
    if v_label[i] == sample['label'].values[i]:
        num_v_all = num_v_all + 1
    if t_label[i] == sample['label'].values[i]:
        num_t_all = num_t_all + 1
    if a_label[i] == sample['label'].values[i]:
        num_a_all = num_a_all + 1
    if l_label[i] == sample['label'].values[i]:
        num_l_all = num_l_all + 1
print("vader,textblob,affin, labMT의 사용결과는 차례대로 :",num_v_all/len(sample),num_t_all/len(sample),num_a_all/len(sample),num_l_all/len(sample))
```

# 문장단위

```python
from nltk.tokenize import sent_tokenize
from nltk import tokenize
v_label_sen = []
t_label_sen = []
a_label_sen = []
num = 1
for i in sample_cleansing:
    para = tokenize.sent_tokenize(i) #문장단위로 자른거 ->리스트형태
    num_v = 0
    num_t = 0
    num_a = 0
    for j in para:
        num_v = num_v + sid.polarity_scores(j)['compound']
        num_t = num_t + TextBlob(j).sentiment.polarity
        num_a = num_a + afn.score(j)
    if num_v>0:
        v_label_sen.append(1)
    else:
        v_label_sen.append(0)
    if num_t>0:
        t_label_sen.append(1)
    else:
        t_label_sen.append(0)
    if num_a>0:
        a_label_sen.append(1)
    else:
        a_label_sen.append(0)
    num = num + 1
num_v_sen = 0
num_t_sen = 0
num_a_sen = 0


for i in range(len(sample)):
    if v_label_sen[i] == sample['label'].values[i]:
        num_v_sen = num_v_sen + 1
    if t_label_sen[i] == sample['label'].values[i]:
        num_t_sen = num_t_sen + 1
    if a_label_sen[i] == sample['label'].values[i]:
        num_a_sen = num_a_sen + 1
        
print("문장 단위로 계산했을때 vader,textblob,afinn의 정확도는 각각 ",num_v_sen/len(sample),num_t_sen/len(sample),num_a_sen/len(sample))
```
