```python
def statics(df):
    means = pd.DataFrame(np.mean(df), columns = ['mean'])
    std = pd.DataFrame(np.std(df), columns = ['std'])
    median = pd.DataFrame(np.median(df, axis = 0), columns = ['median'])
    median.index = df.columns
    q25 = pd.DataFrame(np.percentile(df, 25, axis = 0), columns = ['q25'])
    q25.index = df.columns
    q75 = pd.DataFrame(np.percentile(df, 75, axis = 0), columns = ['q75'])
    q75.index = df.columns
    mins = pd.DataFrame(np.min(df, axis = 0), columns = ['min'])
    maxs = pd.DataFrame(np.max(df, axis = 0), columns = ['max'])
    
    return pd.concat([means, std, mins, q25, median, q75,  maxs], axis = 1)
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
def make_lists(filename):
    df = pd.read_csv(filename)
    
    content_lists = []
    for i in (df['title'] + ' '+df['text_context']):
        i = i.replace('\n', '')
        content_lists.append(i.lower())
    return content_lists

def maketoken(filename):
    contents = make_lists(filename)
    result = []
    for content in contents: 
        temp_text = nltk.word_tokenize(content)
        result.append(len(temp_text))
    print('done!')
    return result
```

```python
at =  maketoken('adhd_origin.csv')
bt=  maketoken('bipolar_origin.csv')
ant =  maketoken('anxiety_origin.csv')
dt =  maketoken('depression_origin.csv')
st =  maketoken('schizo_origin.csv')
nt =  maketoken('non_mh_origin.csv')
```

```python
data = pd.DataFrame({
    'subreddit': 'ADHD',
    'time': 'present',
    '%': a_time['present']*100/at}).append(pd.DataFrame({
    'subreddit': 'ADHD',
    'time': 'past',
    '%': a_time['past']*100/at})).append(pd.DataFrame({
    'subreddit': 'ADHD',
    'time': 'future',
    '%': a_time['future']*100/at}))

data = data.append(pd.DataFrame({
    'subreddit': 'bipolar',
    'time': 'present',
    '%': b_time['present']*100/bt})).append(pd.DataFrame({
    'subreddit': 'bipolar',
    'time': 'past',
    '%': b_time['past']*100/bt})).append(pd.DataFrame({
    'subreddit': 'bipolar',
    'time': 'future',
    '%': b_time['future']*100/bt}))

data = data.append(pd.DataFrame({
    'subreddit': 'anxiety',
    'time': 'present',
    '%': an_time['present']*100/ant})).append(pd.DataFrame({
    'subreddit': 'anxiety',
    'time': 'past',
    '%': an_time['past']*100/ant})).append(pd.DataFrame({
    'subreddit': 'anxiety',
    'time': 'future',
    '%': an_time['future']*100/ant}))
plt.ylim(0,100)
sns.boxplot(data=data, x = 'subreddit', y = '%', hue='time')
```
