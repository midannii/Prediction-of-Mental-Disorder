```python
# same for describe()
def statics(df):
    means = pd.DataFrame(np.mean(df), columns = ['mean'])
    std = pd.DataFrame(np.std(df), columns = ['std'])
    median = pd.DataFrame(np.median(df, axis = 0), columns = ['median'])
    median.index = df.columns
    q25 = pd.DataFrame(np.percentile(df, 25, axis = 0), columns = ['q25'])
    q25.index = df.columns
    q75 = pd.DataFrame(np.percentile(df, 75, axis = 0), columns = ['q75'])
    q75.index = df.columns
    mins = pd.DataFrame(np.min(df, axis = 0), columns = ['min'])
    maxs = pd.DataFrame(np.max(df, axis = 0), columns = ['max'])
    
    return pd.concat([means, std, mins, q25, median, q75,  maxs], axis = 1)
```

```python
def normal_check(col, alpha):
    name_list = ['ADHD', 'bipolar', 'anxiety', 'depression', 'schizo', 'nonMH']
    stas_list = []
    p_list = []
    result = []
    
    stat, p = stats.normaltest(a_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')
        
    stat, p = stats.normaltest(b_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')    
    
    stat, p = stats.normaltest(an_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')
    
    stat, p = stats.normaltest(d_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')
    
    stat, p = stats.normaltest(s_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')
    
    stat, p = stats.normaltest(n_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<alpha: result.append('reject Ho')
    else: result.append('accept Ho')
    
    result = pd.DataFrame({
        'subreddits': name_list, 'statistics': stas_list, 'p-value': p_list, 'result': result
    })
    return result
```

### Kruskal-Wallis H-test

#### Ho: population median of all of the groups are equal.

```python
from scipy import stats
```

```python
def kw(col):
    stas_list = []
    p_list = []
    name_list = ['all subreddits', 'all MH', 'MH & non-MH']
    result = []
    
    stat, p = stats.kruskal(a_alike[col], b_alike[col], an_alike[col], s_alike[col], d_alike[col], n_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
        
    stat, p = stats.kruskal(a_alike[col], b_alike[col], an_alike[col], s_alike[col], d_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')    
    
    mh = list(a_alike[col]) + list(b_alike[col]) + list(an_alike[col]) + list(s_alike[col]) + list(d_alike[col]) 
    stat, p = stats.kruskal(mh, n_alike[col])
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
        
    result = pd.DataFrame({
        'subreddits': name_list, 'statistics': stas_list, 'p-value': p_list, 'result': result
    })
    
    return result
```

```python
## bipolar, depress & schizo , anxiety, ADHD & nonMH

## bonferroni correction: a = 0.05/6
def mwu(col):
    new_a = 0.05/6
    p_list = []
    stas_list = []
    name_list = ['[ADHD & nonMH]', '[bipolar & nonMH]', '[anxiety & nonMH]', '[depression & nonMH]', '[schizo & nonMH]',
                '[ADHD & bipolar]', '[ADHD & anxiety]', '[ADHD & depression]', '[ADHD & schizo]', 
                '[bipolar & anxiety]', '[bipolar & depression]', '[bipolar & schizo]', 
                 '[anxiety & depression]', '[anxiety & schizo]', '[depression & schizo]']
    result = []
    
    stat, p = stats.mannwhitneyu(a_alike[col], n_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(b_alike[col], n_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(an_alike[col], n_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(d_alike[col], n_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(s_alike[col], n_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
        
    stat, p = stats.mannwhitneyu(a_alike[col], b_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')     
    stat, p = stats.mannwhitneyu(a_alike[col], an_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')     
    stat, p = stats.mannwhitneyu(a_alike[col], d_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(a_alike[col], s_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')    
    
    stat, p = stats.mannwhitneyu(b_alike[col], an_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')     
    stat, p = stats.mannwhitneyu(b_alike[col], d_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(b_alike[col], s_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')   
    
    stat, p = stats.mannwhitneyu(an_alike[col], d_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho')
    stat, p = stats.mannwhitneyu(an_alike[col], s_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho') 
    
    stat, p = stats.mannwhitneyu(d_alike[col], s_alike[col], use_continuity=True, alternative=None)
    stas_list.append(stat)
    p_list.append(p)
    if p<new_a: result.append('reject Ho')
    else: result.append('accept Ho') 
        
    print('New alpha: '+ str(new_a))
    result = pd.DataFrame({
        'subreddits': name_list, 'statistics': stas_list, 'p-value': p_list, 'result': result
    })
    return result 
    
```

```python
def my_plot(col, log= False):
    # means 
    values = [np.mean(a_alike[col]), np.mean(b_alike[col]) , np.mean(an_alike[col]) , np.mean(d_alike[col]) , 
            np.mean(s_alike[col]), np.mean(n_alike[col])]
    subreddits = ['ADHD', 'bipolar', 'anxiety', 'depression', 'schizo', 'nonMH']
    # ste
    yerr = [stats.sem(a_alike[col]), stats.sem(b_alike[col]) , stats.sem(an_alike[col]) , stats.sem(d_alike[col]) , 
            stats.sem(s_alike[col]), stats.sem(n_alike[col])]
    
    colors = sns.color_palette('Set2',len(values))
    plt.bar(x = subreddits, height = values,yerr=yerr, capsize=10, color = colors, alpha = 0.7)
    
    plt.ylabel('rates')
    plt.title('mean of '+col+' with error bar')
    plt.show
```

```python
# expected tables 
dt = np.asarray(data)
dt = dt[:,:-1]
result = []
for i in range(len(dt)):
    temp = []
    for j in range(len(dt[0])):
        temp.append(dt[i][-1]*dt[4][j]/dt[-1][-1])
    result.append(temp)
```

```python
exp = pd.DataFrame( {
    '~ q25': result[0], 'q25 ~ median': result[1], 'median ~ q75':result[2], 'a75 ~': result[3]
}).transpose()
exp.columns = list(data.columns)[:-1]
exp # expected values 
```

```python
data2 = data.transpose().iloc[:-1, :-1]
dt = np.asarray(data2)
exp2 = exp.transpose()
expdt = np.asarray(exp2)
```

```python
# chisquare with scipy
result_chi = []
result_p = []
r = []
cv = stats.chi2.ppf(0.05, df=df)
for i in range(len(expdt)):
    chi, p = stats.chisquare(dt[i], f_exp=expdt[i])
    result_p.append(p)
    result_chi.append(chi)
    if chi > cv: r.append('reject Ho')
    else: r.append('accept Ho')
result = pd.DataFrame({
    'chi-statistics': result_chi, 'p-value': result_p, 'critical value': cv, 'result':r
}, index = data2.index)
result
```
